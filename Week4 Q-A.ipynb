```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the telco churn dataset to python environment
data=pd.read_csv(r'C:\Users\Smitha.s1_oob\OneDrive - GEMS Education\Documents\PERSONAL\DS\Telco-Customer-Churn.csv')
print(data)
```

```python
#printing columns of dataframe
print(data.columns)
```

```python
#Calculate female and male churn count
data1=data[['gender','Churn']]
data2=data1[(data1['gender']=='Female') & (data1['Churn']=='Yes')]
print(data2)
data3=data1[(data1['gender']=='Male') & (data1['Churn']=='Yes')]
print(data3)
n1=data2.count()
n2=data3.count()
print(n1,'\n',n2)
```

```python
#Compare churn count with respect to gender.
import numpy as np
x=np.arange(1)
plt.bar(x,n1,width=0.25,label='Female customers churn')
plt.bar(x+0.25,n2,width=0.25,label='Male customers churn')
plt.legend()
plt.show()
```

```python
#Find out how many female senior citizens there in the dataset
data3=data[['gender','SeniorCitizen']]
datasf=data3[(data3['gender']=='Female') & (data3['SeniorCitizen']==1)]
print(datasf)
datasm=data3[(data3['gender']=='Male') & (data3['SeniorCitizen']==1)]
print(datasm)
```

```python
#plot of female and male senior citizens
import numpy as np
x=np.arange(1)
sf=datasf.count()
sm=datasm.count()
print('Female count=',sf,'\nMale count=',sm)
plt.bar(x,sf,width=0.25,label='Female count')
plt.bar(x+0.25,sm,width=0.25,label='Male count')
plt.legend()
plt.show()
```

```python
#Compare 'tenure' with 'Total Charges'
import seaborn as sns
sns.lineplot(x='tenure',y='TotalCharges',data=data)
plt.show()
```

```python
#Find out counts of contract is preferred by the senior citizen.
data4=data[['SeniorCitizen','Contract']]
print(data4)
c1=data4[(data4['SeniorCitizen']==1)& (data4['Contract']=='Month-to-month')].count()
print(c1)
c2=data4[(data4['SeniorCitizen']==1)& (data4['Contract']=='One year')].count()
print(c2)
c3=data4[(data4['SeniorCitizen']==1)& (data4['Contract']=='Two year')].count()
print(c3)
contract_counts=[c1[0],c2[0],c3[0]]
print(contract_counts)
```

```python
#plot for type of contracts preferred by senior citizen
c=['Month-to-Month','One year','Two year']
plt.pie(contract_counts,labels=c,autopct='%.f%%')
plt.title('Type of contracts preferred by Senior Citizens')
plt.show()
```
