```python
import pandas as pd
data=pd.read_excel('/content/Dataset_w6.xlsx')
print(data)
```

```python
print('No: of countries=',data['country'].count())
```

```python
#Data Visualization
import matplotlib.pyplot as plt
x=data['country']
y1=data['GDP']
y2=data['API']
y3=data['International trade']
plt.subplot(3,1,1)
plt.plot(x,y1,color='r',label='GDP per capita')
plt.subplots_adjust(hspace =0.6,wspace=0.4)
plt.subplot(3,1,2)
plt.plot(x,y2,color='k',label='Agricultural production Index')
plt.subplots_adjust(hspace =0.6,wspace=0.4)
plt.subplot(3,1,3)
plt.plot(x,y3,color='m',label='International Trade:Imports')
plt.show()
```

```python
data1=data[['country','Region']]
d=data1.groupby('Region')['country'].count()
print(d)
```

```python
data2=data.groupby('Region')['GDP'].mean()
print(data2)
print('On average Region with highest GDP=')
print(data2.idxmax(),data2.max())
```

```python
data3=data.groupby('Region')['International trade'].mean()
print(data3)
print('On average Region with International trade=')
print(data3.idxmax(),data3.max())
```

```python
data4=data.groupby('Region')['GDP'].std()
print(data4)
print('Region with less standard deviation(highly consistent GDP)=')
print(data4.idxmin(),data4.min())
```
